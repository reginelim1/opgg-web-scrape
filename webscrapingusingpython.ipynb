```python
content = driver.page_source
soup = BeautifulSoup(content, 'html5lib')
soup.prettify()
#with open('html.txt', 'w') as f:
#    f.write(soup.prettify())
```

```python
import pandas as pd

champion = []
status = []
duration = []
kill = []
death = []
assist = []
kp = []
dkey = []
fkey = []
cs = []
cspm = []
c1 = []
c2 = []
c3 = []
c4 = []
c5 = []
c6 = []
c7 = []
c8 = []
c9 = []
c10 = []
team1 = []
team2 = []
ward = []
multi = []
teamb = []
teamr = []
items1 = []
items2 = []
items3 = []
items4 = []
items5 = []
items6 = []
items7 = []
mvp = []
ace = []

for a in soup.findAll('li', attrs={'class': 'css-1qq23jn e1iiyghw3'}):
  champ = a.img['alt']
  stat = a.find('div', attrs={'class': 'result'})
  dur = a.find('div', attrs={'class': 'length'})
  kd = a.find('div', attrs={'class': 'k-d-a'}).text.split(" /")
  kip = a.find('div', attrs={'class': 'p-kill'})
  d1 = a.select('div > img')[0]['alt']
  f1 = a.select('div > img')[1]['alt']
  cs1 = a.find('div', class_='cs')
  tm1 = a.find('div', class_='participants').select('div > img')[0]['alt']
  tm2 = a.find('div', class_='participants').select('div > img')[1]['alt']
  tm3 = a.find('div', class_='participants').select('div > img')[2]['alt'] 
  tm4 = a.find('div', class_='participants').select('div > img')[3]['alt']
  tm5 = a.find('div', class_='participants').select('div > img')[4]['alt']
  tm6 = a.find('div', class_='participants').select('div > img')[5]['alt'] 
  tm7 = a.find('div', class_='participants').select('div > img')[6]['alt']
  tm8 = a.find('div', class_='participants').select('div > img')[7]['alt']
  tm9 = a.find('div', class_='participants').select('div > img')[8]['alt']
  tm10 = a.find('div', class_='participants').select('div > img')[9]['alt']
  ward1 = a.find('div', class_='ward')
  mk = a.find('div', attrs={'class': 'multi-kill'})
  if mk == None:
    mk = 'None'
  else:
    mk = mk.text
  
  try:
    it1 = a.find('div', class_='items').select('ul > li > div > img')[0]['alt']
    it2 = a.find('div', class_='items').select('ul > li > div > img')[1]['alt']
    it3 = a.find('div', class_='items').select('ul > li > div > img')[2]['alt']
    it4 = a.find('div', class_='items').select('ul > li > div > img')[3]['alt']
    it5 = a.find('div', class_='items').select('ul > li > div > img')[4]['alt']
    it6 = a.find('div', class_='items').select('ul > li > div > img')[5]['alt']
  except:
    if len(it1) < 1:
      it1 = 'None'
      it2 = 'None'
      it3 = 'None'
      it4 = 'None'
      it5 = 'None'
      it6 = 'None'
    elif len(it2) < 1:
      it2 = 'None'
      it3 = 'None'
      it4 = 'None'
      it5 = 'None'
      it6 = 'None'
    elif len(it3) < 1:
      it3 = 'None'
      it4 = 'None'
      it5 = 'None'
      it6 = 'None'
    elif len(it4) < 1:
      it4 = 'None'
      it5 = 'None'
      it6 = 'None'
    elif len(it5) < 1:
      it5 = 'None'
      it6 = 'None'
    else:
      it6 = 'None'

  it7 = a.find('div', class_='items').find('div', class_='ward').find('img')['alt']
  mvp1 = a.find('div', class_='badge mvp')
  ace1 = a.find('div', class_='badge ace')
  if mvp1 != None:
    mvp1 = "True"
    ace1 = "False"
  elif ace1 != None:
    ace1 = "True"
    mvp1 = "False"
  else:
    ace1 = "False"
    mvp1 = "False"

  champion.append(champ)
  status.append(stat.text)
  duration.append(dur.text.replace('m ', ':').replace('s',''))
  kill.append(kd[0])
  death.append(kd[1])
  assist.append(kd[2])
  kp.append(kip.text.split()[1].strip("%\n "))
  dkey.append(d1)
  fkey.append(f1)
  cs.append(cs1.text.split()[1])
  cspm.append(cs1.text.split()[2].split('(', 1)[1].split(')')[0])
  c1.append(tm1)
  c2.append(tm2)
  c3.append(tm3)
  c4.append(tm4)
  c5.append(tm5)
  c6.append(tm6)
  c7.append(tm7)
  c8.append(tm8)
  c9.append(tm9)
  c10.append(tm10)
  ward.append(ward1.text.split()[2])
  multi.append(mk)
  items7.append(it7)
  items1.append(it1)
  items2.append(it2)
  items3.append(it3)
  items4.append(it4)
  items5.append(it5)
  items6.append(it6)
  mvp.append(mvp1)
  ace.append(ace1)
  
  teamb.append([tm1, tm2, tm3, tm4, tm5])
  teamr.append([tm6, tm7, tm8, tm9, tm10])

  if (tm1 == champ) or (tm2 == champ) or (tm3 == champ) or (tm4 == champ) or (tm5 == champ):
    team1 = teamb
    team2 = teamr
  else:
    team1 = teamr
    team2 = teamb


df = pd.DataFrame({'Champion': champion, 'Status': status, 'Duration': duration, 'Kill': kill, 'Death': death, 'Assist': assist,
                   'Kill Participation (%)': kp, 'Multikill': multi, 'MVP': mvp, 'Ace': ace,
                   'D': dkey, 'F': fkey, 'CS': cs, 'CS/M': cspm, 
                   "Champ1": c1, "Champ2": c2, "Champ3": c3, "Champ4": c4, "Champ5": c5, "Champ6": c6, "Champ7": c7, "Champ8": c8, "Champ9": c9, "Champ10": c10,
                   'Blue': teamb, 'Red': teamr, "Opponent": team2,
                   "Ward": ward, "Ward Type": items7, "Item 1": items1, "Item 2": items2, "Item 3": items3, "Item 4": items4, "Item 5": items5, "Item 6": items6
                   })

df['Duration'] = pd.to_datetime(df.Duration, format="%M:%S").dt.time
df[['Kill', 'Death', 'Assist', 'CS', 'CS/M', 'Kill Participation (%)']] = df[['Kill', 'Death', 'Assist', 'CS', 'CS/M', 'Kill Participation (%)']].apply(pd.to_numeric)

pd.set_option('display.max_columns', None)
df.head(10)
```
