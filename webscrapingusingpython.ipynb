```python
content = driver.page_source
soup = BeautifulSoup(content, 'html5lib')
soup.prettify()
with open('html.txt', 'w') as f:
    f.write(soup.prettify())
```

```python
import pandas as pd

df = []

for a in soup.findAll('li', attrs={'class': 'css-1qq23jn e1iiyghw3'}):
  df1 = {}
  df1["Queue Type"] = a.find("div", class_="type").text
  df1["Champion"] = champ = a.img['alt']
  df1["Status"] = a.find('div', attrs={'class': 'result'}).text
  df1["Duration"] = a.find('div', attrs={'class': 'length'}).text.replace('m ', ':').replace('s','')
  kd = a.find('div', attrs={'class': 'k-d-a'}).text.split(" /")
  df1["Kill"] = kd[0]
  df1["Death"] = kd[1]
  df1["Assist"] = kd[2]
  df1["Kill Participation (%)"] = a.find('div', attrs={'class': 'p-kill'}).text.split()[1].strip("%\n ")
  
  mk = a.find('div', attrs={'class': 'multi-kill'})
  if mk == None:
    df1["Multikill"] = 'None'
  else:
    df1["Multikill"] = mk.text
  
  mvp1 = a.find('div', class_='badge mvp')
  ace1 = a.find('div', class_='badge ace')
  if mvp1 != None:
    df1["MVP"] = "True"
    df1["Ace"] = "False"
  elif ace1 != None:
    df1["Ace"] = "True"
    df1["MVP"] = "False"
  else:
    df1["Ace"] = "False"
    df1["MVP"] = "False"
  
  df1["D"] = a.select('div > img')[0]['alt']
  df1["F"] = a.select('div > img')[1]['alt']
  cs1 = a.find('div', class_='cs')
  df1["CS"] = cs1.text.split()[1]
  df1["CS/M"] = cs1.text.split()[2].split('(', 1)[1].split(')')[0]
  df1["Champ1"] = tm1 = a.find('div', class_='participants').select('div > img')[0]['alt']
  df1["Champ2"] = tm2 = a.find('div', class_='participants').select('div > img')[1]['alt']
  df1["Champ3"] = tm3 = a.find('div', class_='participants').select('div > img')[2]['alt'] 
  df1["Champ4"] = tm4 = a.find('div', class_='participants').select('div > img')[3]['alt']
  df1["Champ5"] = tm5 = a.find('div', class_='participants').select('div > img')[4]['alt']
  df1["Champ6"] = tm6 = a.find('div', class_='participants').select('div > img')[5]['alt'] 
  df1["Champ7"] = tm7 = a.find('div', class_='participants').select('div > img')[6]['alt']
  df1["Champ8"] = tm8 = a.find('div', class_='participants').select('div > img')[7]['alt']
  df1["Champ9"] = tm9 = a.find('div', class_='participants').select('div > img')[8]['alt']
  df1["Champ10"] = tm10 = a.find('div', class_='participants').select('div > img')[9]['alt']

  df1["Blue"] = teamb = [tm1, tm2, tm3, tm4, tm5]
  df1["Red"] = teamr = [tm6, tm7, tm8, tm9, tm10]

  if (tm1 == champ) or (tm2 == champ) or (tm3 == champ) or (tm4 == champ) or (tm5 == champ):
    team1 = teamb
    df1["Opponent"] = teamr
  else:
    team1 = teamr
    df1["Opponent"] = teamb
  
  df1["Control Ward"] = a.find('div', class_='ward').text.split()[2]
  df1["Ward Type"] = a.find('div', class_='items').find('div', class_='ward').find('img')['alt']
  
  try:
    df1["Item 1"] = it1 = a.find('div', class_='items').select('ul > li > div > img')[0]['alt']
    df1["Item 2"] = it2 = a.find('div', class_='items').select('ul > li > div > img')[1]['alt']
    df1["Item 3"] = it3 = a.find('div', class_='items').select('ul > li > div > img')[2]['alt']
    df1["Item 4"] = it4 = a.find('div', class_='items').select('ul > li > div > img')[3]['alt']
    df1["Item 5"] = it5 = a.find('div', class_='items').select('ul > li > div > img')[4]['alt']
    df1["Item 6"] = it6 = a.find('div', class_='items').select('ul > li > div > img')[5]['alt']
  except:
    if len(it1) < 1:
      df1["Item 1"] = 'None'
      df1["Item 2"] = 'None'
      df1["Item 3"] = 'None'
      df1["Item 4"] = 'None'
      df1["Item 5"] = 'None'
      df1["Item 6"] = 'None'
    elif len(it2) < 1:
      df1["Item 2"] = 'None'
      df1["Item 3"] = 'None'
      df1["Item 4"] = 'None'
      df1["Item 5"] = 'None'
      df1["Item 6"] = 'None'
    elif len(it3) < 1:
      df1["Item 3"] = 'None'
      df1["Item 4"] = 'None'
      df1["Item 5"] = 'None'
      df1["Item 6"] = 'None'
    elif len(it4) < 1:
      df1["Item 4"] = 'None'
      df1["Item 5"] = 'None'
      df1["Item 6"] = 'None'
    elif len(it5) < 1:
      df1["Item 5"] = 'None'
      df1["Item 6"] = 'None'
    else:
      df1["Item 6"] = 'None'
 
  df.append(df1)

df = pd.DataFrame(df)
df['Duration'] = pd.to_datetime(df.Duration, format="%M:%S").dt.time

pd.set_option('display.max_columns', None)
df.head(10)
```

```python
newtab = []
for a in soup.findAll('li', attrs={'class': 'css-1qq23jn e1iiyghw3'}):
  rows = a.find("table").find("tbody").findAll("tr")
  for row in rows:
    newtable = {}
    cols = row.findAll("td")
    cols[3].get_text(" ", strip = True)
    newtable["Summoner Name"] = [text for text in cols[3].stripped_strings][0]
    try:
      newtable["Damage Dealt"] = int((cols[6].find("div", class_="dealt").text).replace(',',''))
      newtable["Damage Taken"] = int((cols[6].find("div", class_="taken").text).replace(',',''))
    except:
      newtable["Damage Dealt"] = int((cols[5].find("div", class_="dealt").text).replace(',',''))
      newtable["Damage Taken"] = int((cols[5].find("div", class_="dealt").text).replace(',',''))
    cols[7].get_text(" ", strip = True)
    try:
      newtable["Wards Placed"] = ([text for text in cols[7].stripped_strings][1]).split(" / ")[0]
      newtable["Wards Killed"] = ([text for text in cols[7].stripped_strings][1]).split(" / ")[1]
    except:
      newtable["Wards Placed"] = ([text for text in cols[6].stripped_strings][1]).split(" / ")[0]
      newtable["Wards Killed"] = ([text for text in cols[6].stripped_strings][1]).split(" / ")[1]
    newtab.append(newtable)



dfdam = pd.DataFrame(newtab)
dfdam = dfdam[dfdam.groupby("Summoner Name")["Summoner Name"].transform('count') == df.shape[0]].reset_index(drop=True).drop(columns=["Summoner Name"])
df = pd.concat([df, dfdam], join="inner", axis = 1)
df[['Kill', 'Death', 'Assist', 'CS', 'CS/M', 'Kill Participation (%)', 'Damage Dealt', 'Damage Taken', 'Wards Placed', 'Wards Killed']] = df[['Kill', 'Death', 'Assist', 'CS', 'CS/M', 'Kill Participation (%)', 'Damage Dealt', 'Damage Taken', 'Wards Placed', 'Wards Killed']].apply(pd.to_numeric)
df
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict

label = df[['Status']]
main = df.drop(df[(df["Queue Type"] != "Ranked Solo")].index, inplace=True)
main = df.drop(columns=['Status', 'D', 'F', 'Champ1', 'Champ2', 'Champ3', 'Champ4', 'Champ5', 'Champ6', 'Champ7', 'Champ8', 'Champ9', 'Champ10', 'Queue Type'])
#print(list(main.columns))

#Label encoding

encolist = ['Duration', 'Champion', 'Multikill', 'MVP', 'Ace', 'Opponent', 'Blue', 'Red', 'Item 1', 'Item 2', 'Item 3', 'Item 4', 'Item 5', 'Item 6', 'Ward Type']
main['Blue'] = main['Blue'].apply(lambda x: ','.join(map(str, x)))
main['Opponent'] = main['Opponent'].apply(lambda x: ','.join(map(str, x)))
main['Red'] = main['Red'].apply(lambda x: ','.join(map(str, x)))
label_encoders = {}
for x in encolist:
  label_encoders[x] = preprocessing.LabelEncoder()
  main[x] = label_encoders[x].fit_transform(main[x])


X_train, X_test, y_train, y_test = train_test_split(main, label, test_size=0.2, random_state=507)

model = tree.DecisionTreeClassifier(random_state=507)
model.fit(X_train, y_train)
print(classification_report(y_test, model.predict(X_test)))
print('Accuracy tts', accuracy_score(y_test, model.predict(X_test)))
print(model.predict(X_test))
tree.plot_tree(model.fit(X_train, y_train))
cvpred = cross_val_predict(model, main, label, cv=5)
print(cvpred)
print('cv', (cross_val_score(model, main, label, cv=5, scoring='accuracy')).mean())
print(classification_report(label, cvpred))
print('Accuracy cv', accuracy_score(label, cvpred))
```
